```python
import numpy as np
a =[[[1,2,3],[2,3,4],[4,5,6]],[[1,2,3],[2,3,4],[4,5,6]]]
b = np.array(a)
print(b.shape) #数组维度
print(b.ndim)#数组秩  维度的长度
print(b.size)#数组大小  维度元素的乘积
print(b.dtype)#用来描述数组中元素类型的对象
print(b.data)#包含实际数组元素的缓冲区
print(b.itemsize)#数组中每个元素的字节大小。
#例如，一个元素类型为float64的数组itemsiz属性值为8(=64/8),
#又如，一个元素类型为complex32的数组item属性为4(=32/8).
print(np.argmin(b,axis=2))#最大值得索引
print(b.argmax(axis=1))
print(b.max())#最大值、
print(b.mean())#均值
print(b.flatten())#展开为一维数组
print(b.ravel())#可以解析的一维数组
```

```python
import numpy as np

e = np.ones((3,3))#创建3X3的2维数组，出初始化所有元素为1
print(e)
print(e.dtype)

f = np.repeat(3,4)#创建一维数组，元素值把3重复4次
print(f)
g = np.zeros((2,2,3),dtype=np.uint8)#创建3维数组初始化所有元素值为0
print(g)
print(g.shape)
h =g.astype(np.float)#用另一种类型表示
print(h)
l = np.arange(10)
print(l)
m = np.linspace(0,6,5)#等差数列0,6之间取5个值
print(m)
np.save('g.npy',g)
q = np.load('g.npy')
print(q)
```

```python
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['xtick.labelsize'] = 24
mpl.rcParams['ytick.labelsize'] = 24
np.random.seed(42)
x = np.linspace(0,5,100)

y = 2*np.sin(x) + 0.3*x**2
y_data = y + np.random.normal(scale=0.3,size= 100)

plt.figure('data')

plt.plot(x,y_data,'.')

plt.figure('modal')
plt.plot(x,y)
plt.figure("data&modal")
plt.plot(x,y,'k',lw=3)
plt.scatter(x,y_data)
plt.show()
```
